# Exercise 2.2

Evaluate the similarities of the homeworks of the students.

At a homework level, then as a student level.


```python
#Se define función para medir distancia con jaccard, dado que hay datos nulos no se puede utilizar distancia de coseno
def jaccard_similarity(query, document):
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    if len(union) == 0:
        return 0  # Manejar el caso de la división por cero
    return len(intersection) / len(union)
```

```python
data_trans = data_trans.fillna('')
```

```python
# Iterar sobre cada tarea y calcular las similitudes entre todos los estudiantes (filas)
for i in range(1, len(data_trans.columns)):
    task = data_trans.columns[i]
    print(f"Comparando '{task}' con todas las demás tareas y todas las filas:")
    similarities = []
    for index, row in data_trans.iterrows():
        current_task = row[i]
        for j, row2 in data_trans.iterrows():
            if index != j: 
                comparison_task = row2[i]
                similarity = jaccard_similarity(current_task.split(), comparison_task.split())
                similarities.append((index, j, similarity))

    # Ordenar las similitudes y obtener las más altas
    sorted_similarities = sorted(similarities, key=lambda x: x[2], reverse=True)[:3]
    # Imprime 3 más altas
    for idx1, idx2, similarity in sorted_similarities:
        print(f"Similitud entre estudiante {idx1} y estudiante {idx2}: {similarity}")
```

```python
#Similaridades para estudiantes
# Calcula y muestra la similitud de Jaccard entre todas las filas
similarities_estud = []
for i, (index, row) in enumerate(data_trans.iterrows()):
    for j, (index2, row2) in enumerate(data_trans.iterrows()):
        if i < j:  # Solo calcula la similitud una vez por par de filas
            similarity = jaccard_similarity(row[1:], row2[1:])
            similarities_estud.append((index, index2, similarity))
            print(f"Similitud entre estudiante {index} y estudiante {index2}: {similarity}")
```

```python
#Top 3 de estudiantes similares - JACCARD
# Ordenar la lista de similitudes por el índice de similitud en orden descendente
similarities_estud.sort(key=lambda x: x[2], reverse=True)

# Mostrar el top 3 de similitudes más altas
print("Top 3 de similitudes más altas:")
for i in range(min(3, len(similarities_estud))):
    print(f"Similitud entre estudiante {similarities_estud[i][0]} y estudiante {similarities_estud[i][1]}: {similarities_estud[i][2]}")
```

Classify group members into different subgroups (minimum 3) according to a characteristic of your choice.

```python
#Se dividirán en grupo de estudiantes que sólo usan mayúculas, estudiantes que usan puntuación y estudiantes que no
# Concatenamos todas las columnas de texto para cada fila
data_trans['Text'] = data_trans.apply(lambda row: ' '.join([str(row[column]) for column in data_trans.columns if column != 'Genero']), axis=1)
```

```python
# usa o no puntuación
data_trans['Puntuacion'] = data_trans['Text'].str.contains(r'[.,!?;:]').map({True: 'si', False: 'no'})
```

```python
# escribe todo con mayúculas
data_trans['mayus'] = data_trans['Text'].str.isupper().map({True: 'si', False: 'no'})
```

```python
# grupos basados en las 3 categorías
def clasifica(row):
    if row['Puntuacion'] == 'si':
        return 'Usa puntuación'
    elif row['mayus'] == 'si':
        return 'Sólo usa mayus'
    else:
        return 'No puntuación o mayúsculas'

data_trans['Group'] = data_trans.apply(clasifica, axis=1)
```

```python
# Contar cuántos estudiantes hay en cada grupo
Grupo1 = (data_trans['Puntuacion'] == 'si').sum()
Grupo2 = (data_trans['mayus'] == 'si').sum()
Grupo3 = len(data_trans) - Grupo1 - Grupo2

# Mostrar los resultados
print("Número de estudiantes engrupo 1:", Grupo1)
print("Número de estudiantes engrupo 2:", Grupo2)
print("Número de estudiantes engrupo 3:", Grupo3)
```
